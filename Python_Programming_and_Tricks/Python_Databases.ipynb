```python
import MySQLdb
import MySQLdb.cursors

db = MySQLdb.connect(host='localhost',user='root',passwd='',db='douban',port=3306,charset='utf8',cursorclass=MySQLdb.cursors.DictCursor)
db.autocommit(True)
cursor = db.cursor()

fr = open('C:/Users/tianjiayang/doubanmovie/movie crawler and visualization/crawler/clean_data.txt','r',encoding='utf8')

count = 0
for line in fr:
    count += 1
    print(count)
    if count == 1:
        continue
    line = line.strip().split('^')
    cursor.execute("insert into movie(title,url,score,length,description) values(%s,%s,%s,%s,%s)",[line[1],line[2],line[4],line[-3],line[-1]])
    print(count,'done')




fr.close()
```
